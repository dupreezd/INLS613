```python
             
privacyCaseSet = set([])
#for each row in the issues column of issues csv
issuesFile = open("database.csv")
issuesReader = csv.reader(issuesFile, delimiter=',')
for row in issuesReader:
    if row[39] == 30040.0 or 50010.0 or 50020.0 or 50030.0 or 500040.0:
    #if the integers value matches 30040, 50010, 50020, 50030, or 50040 (floats)
        privacyCaseSet.add(row[0])
        #add case_id value of same row to set
print(privacyCaseSet)
```

```python
num_features = 250
tf_vectorizer = CountVectorizer(max_df=25, min_df= 5 , max_features=num_features, stop_words='english')
```

```python
n_samples = 200
data_samples = random.sample(test.text.tolist(), n_samples)
tf_data_samples = tf_vectorizer.fit_transform(data_samples) # Tokenizing and getting the feature counts. 
tf_feature_names = tf_vectorizer.get_feature_names()
tf_data_samples
```

```python
num_topics = 15 # Tune this parameter. Also consider tuning the parameters max_iter, learning_offset, and random_state. 

lda = LatentDirichletAllocation(n_components=num_topics, max_iter=500, learning_method='online', learning_offset=50.,random_state=1).fit(tf_data_samples)
lda.score(tf_data_samples) # Log-likelihood score. Higher the better (magnitude). This basically mean, given the parameters we have chosen, how best are we able to explain the data. 
```

```python
def print_topics(model, vectorizer, top_n=5):    #### This is the code to print the topics. top_n can be changed. 
    for idx, topic in enumerate(model.components_):
        print("Topic %d:" % (idx))
        print([(vectorizer.get_feature_names()[i], topic[i])
                        for i in topic.argsort()[:-top_n - 1:-1]])
print_topics(lda,tf_vectorizer)
```

```python
doc_topic_dist = lda.transform(tf_data_samples) ## Getting the topic distribution in each document. 
```
