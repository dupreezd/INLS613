```python
privacyCaseSet = set([])
#for each row in the issues column of issues csv
issuesFile = open("database.csv")
issuesReader = csv.reader(issuesFile, delimiter=',')
for row in issuesReader:
    if row[39] == 30040.0 or 50010.0 or 50020.0 or 50030.0 or 500040.0:
    #if the integers value matches 30040, 50010, 50020, 50030, or 50040 (floats)
        privacyCaseSet.add(row[0])
        #add case_id value of same row to set
print(privacyCaseSet)
```

```python
csv.field_size_limit(100000000)
opinionsFile = open("all_opinions.csv")
opinionsReader = csv.reader(opinionsFile, delimiter=',')
dictionary = {}
count = 0
for r in opinionsReader:
    if r[9] in privacyCaseSet:
        #count += 1
        
        if r[9]+'-01' in dictionary.keys():
            if r[9]+'-02' in dictionary.keys():
                if r[9]+'-03' in dictionary.keys():
                    if r[9]+'-04' in dictionary.keys():
                        dictionary[r[9]+'-05'] = r[-1]
                    else: dictionary[r[9]+'-04'] = r[-1]
                else: dictionary[r[9]+'-03'] = r[-1]
            else: dictionary[r[9]+'-02'] = r[-1]
        else: dictionary[r[9]+'-01'] = r[-1]           
print(len(dictionary))
print(count)
```

```python
with open('dict.csv', 'w') as csv_file:
    writer = csv.writer(csv_file)
   # writer.writerow(['id','text'])
    for key, value in dictionary.items():
        writer.writerow([key, value])
```

```python
from sklearn import preprocessing
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import random
```

```python
num_features = 100 # you can change this to which ever value you think is right. 
tf_vectorizer = CountVectorizer(max_df=.2, min_df=5, max_features=num_features, stop_words='english')
```

```python
from sklearn.decomposition import LatentDirichletAllocation ### Importing the LDA library
```

```python
n_samples = 3000
data_samples = random.sample(text.text.tolist(), n_samples)
```

```python
tf_data_samples = tf_vectorizer.fit_transform(data_samples) # Tokenizing and getting the feature counts. 
tf_feature_names = tf_vectorizer.get_feature_names()
```

```python
tf_data_samples ## This is a matrix
```

```python
num_topics = 20 # Tune this parameter. Also consider tuning the parameters max_iter, learning_offset, and random_state. 
#num_topics2 = 100   #40 was best so far 2720000     -2754702.6763839168 for this set
#num_topics3 = 150
lda = LatentDirichletAllocation(n_components=num_topics, max_iter=100, learning_method='online', learning_offset=50.,random_state=1).fit(tf_data_samples)
```

```python
lda.score(tf_data_samples) # Log-likelihood score. Higher the better (magnitude). This basically mean, given the parameters we have chosen, how best are we able to explain the data. 
```

```python
def print_topics(model, vectorizer, top_n=8):    #### This is the code to print the topics. top_n can be changed. 
    for idx, topic in enumerate(model.components_):
        print("Topic %d:" % (idx))
        print([(vectorizer.get_feature_names()[i], topic[i])
                        for i in topic.argsort()[:-top_n - 1:-1]])
```

```python
print_topics(lda,tf_vectorizer) ### Printing all the topics, and the top 5 words in them. 
```
